```python
import numpy; import numpy as np

class MDP:
    def __init__(self, nstates, nactions, nfeatures, rewardProp=0.2, randomness=0.4, type='terminal'):
        if type=='terminal':
            self.P = np.random.random((nstates,nactions,nstates)) ** (1./randomness)
            self.P /= np.sum(self.P, axis=2)[:,:,None]
            self.r = np.zeros((nstates))
            r = 1
            for i in range(int(rewardProp*nstates)): 
                self.r[np.random.randint(0,nstates)] = r
                r /= 2.
            print self.r
            self.initialState = np.argmin(self.r) # some state with no reward
            self.terminal = lambda x,r: r>0
        elif type == 'manychain':
            self.P = np.zeros((nstates,nactions,nstates))
            nsteps = nstates/nactions
            for i in range(nsteps-1):
                for j in range(nactions):
                    for k in range(nactions):
                        self.P[i*nactions+j,:,(i+1)*nactions+k] = np.random.random((nactions))**(1./randomness)
            self.P /= np.sum(self.P, axis=2)[:,:,None]
            self.r = np.random.random(nstates) ** (1./randomness)
            self.initialState = 0
            self.terminal = lambda s,r: s > (nsteps-1)*nactions
        self.x = np.eye(nstates)
        #self.x = np.random.random((nstates,nfeatures))
        self.nstates = nstates; self.nactions = nactions; self.nfeatures = nfeatures
        
    def episode(self):
        s = self.initialState
        a = 0
        r = 0
        isnt = 0
        while True:
            a = yield (s,r,isnt)
            if a == 'terminate': break
            s = np.random.multinomial(1,self.P[s,a]).argmax()
            r = self.r[s]
            isnt = float(not self.terminal(s,r))

def softmax(x): s = np.exp(x); return s/s.sum()
            
def AC(mdp, alr, clr, gamma=0.25, nsteps=10000):
    V = np.random.uniform(-1,1,(mdp.nfeatures,))
    W = np.random.uniform(-1,1,(mdp.nfeatures,mdp.nactions))
    ep = mdp.episode()
    s, r, _ = ep.next()
    R = 0
    Rs = []
    Vdeltas = []
    Wdeltas = []
    tderrors = []
    nablapio = []
    for i in range(1,nsteps):
        x = mdp.x[s]
        pol = softmax(np.dot(x,W))
        a = np.random.multinomial(1,pol).argmax()
        sp, r, isnonterminal = ep.send(a)
        
        Vs = np.dot(x,V)
        Vsp = np.dot(mdp.x[sp],V) * isnonterminal
        # critic update
        V = V + clr * (r + gamma * Vsp - Vs) * x
        Vdeltas.append(np.mean(abs((r + gamma * Vsp - Vs) * x)))
        tderrors.append((r + gamma * Vsp - Vs))
        # actor update
        nabla_o_pia = -pol*pol[a]; nabla_o_pia[a] = pol[a] * (1-pol[a])
        nabla_W_pia = np.outer(x, nabla_o_pia)
        W = W + alr * (r + gamma * Vsp - Vs) * nabla_W_pia  / pol[a]
        Wdeltas.append(np.mean(abs((r + gamma * Vsp - Vs) * nabla_W_pia  / pol[a])))
        nablapio.append(nabla_o_pia)
        s = sp
        R += r
        if not isnonterminal:
            ep = mdp.episode()
            s,r,_ = ep.next()
            Rs.append(R)
            R = 0
            if len(Rs) > 500 and np.mean(Rs[-500:]) == 1:
                break
    return map(numpy.float32,[Rs,Vdeltas,Wdeltas,tderrors,nablapio])

        
        
```

```python
def plotmeans(x,nbins=50,**kwargs):
    nbins = min(nbins,x.shape[0])
    n = x.shape[0]/nbins
    N = x.shape[0]/n
    n = x.shape[0]/N
    #N = x.shape[0]-x.shape[0]%n
    pp.plot(x[:N*n].reshape((N,n)).mean(axis=1),**kwargs)
```
